```python
plt.close('all')
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
```

```python
fig, axis = plt.subplots(1,2,figsize=(16,4))
axis[0].plot(data,'bo--', drawstyle="steps-post", label='steps-post')
axis[0].plot(data,'bo--', label='Default')
axis[1].plot(data, color="b", linestyle="dashed", marker="o", drawstyle="steps-post", label='steps-post')
axis[1].plot(data, color="b", linestyle="dashed", marker="o", label='Default')
axis[0].legend(loc='best')
axis[1].legend(loc='best')
plt.xlim()
```

```python
from datetime import datetime
fig = plt.figure()
axis = fig.add_subplot(1,1,1)
data = pd.read_csv('../pydata-book/examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=axis, style='k--')

crisis_data = [
    (datetime(2007,10,11), 'Peak of bull market'),
    (datetime(2008,3,12), 'Bear Stearns Fails'),
    (datetime(2008,9,15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    axis.annotate(label, xy=(date, spx.asof(date) + 75),
                 xytext=(date,spx.asof(date)+225),
                 arrowprops=dict(facecolor='black',headwidth=4,width=2, headlength=4),
                 horizontalalignment='left', verticalalignment='top')

    axis.set_xlim(['1/1/2007', '1/1/2011'])
    axis.set_ylim([600,1800])
    axis.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
fig, axes = plt.subplots(1,3, figsize=(16,4))
ax1,ax2,ax3 = axes

rect = plt.Rectangle((0.2,0.75), 0.4, 0.15, color='k', alpha=0.3)
ax1.add_patch(rect)

circ = plt.Circle((0.4, 0.4), 0.2)
ax2.add_patch(circ)

pgon = plt.Polygon([[0.4, 0.2], [0.8,0.8], [0.8,0.2]])
ax3.add_patch(pgon)
```

```python
tips = pd.read_csv('../pydata-book/examples/tips.csv')
tips
```

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts = party_counts.loc[:,2:5]
```

```python
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
party_pcts.plot.bar()
```
